```python
import pandas as pd
```

```python
df = pd.read_excel("../datasets/attractions/New_York_Tourist_Locations.xlsx")
```

```python
df.shape
```

```python
# Split 'Address' on the zip code and keep the part before it
df['Address'] = df['Address'].str.split('(\d{5})').str[0]
for idx, e in df.iterrows():
    if "Manhattan" in e['Address']:
        df.loc[idx, 'borough'] = 'NY'
        df.loc[idx, 'Address'] = df.loc[idx, 'Address'].rsplit("Manhattan")[0]
    elif "New York" in e['Address']:
        df.loc[idx, 'borough'] = 'NY'
        df.loc[idx, 'Address'] = df.loc[idx, 'Address'].rsplit("New York")[0]
    elif "Brooklyn" in e['Address']:
        df.loc[idx, 'borough'] = 'K'
        if "Expressway" in e['Address']:
            df.loc[idx, 'Address'] = df.loc[idx, 'Address'].rsplit("Brooklyn")[0:1]
        else:
            df.loc[idx, 'Address'] = df.loc[idx, 'Address'].rsplit("Brooklyn")[0]
    elif "Bronx" in e['Address']:
        df.loc[idx, 'borough'] = 'BX'
        df.loc[idx, 'Address'] = df.loc[idx, 'Address'].rsplit("Bronx")[0]
    elif "Queens" in e['Address']:
        df.loc[idx, 'borough'] = 'Q'
        df.loc[idx, 'Address'] = df.loc[idx, 'Address'].rsplit("Queens")[0]
    elif "Staten Island" in e['Address']:
        df.loc[idx, 'borough'] = 'R'
        df.loc[idx, 'Address'] = df.loc[idx, 'Address'].rsplit("Staten Island")[0]


df['Address'] = df['Address'].str.lower()

df['Address'] = df['Address'].str.replace('(\d+)(st|nd|rd|th)', r'\1', regex=True)
df['Address'] = df['Address'].str.replace(' ave. ', ' avenue ')
df['Address'] = df['Address'].str.replace(' ave ', ' avenue ')
df['Address'] = df['Address'].str.replace(' ave$', ' avenue', regex=True)
df['Address'] = df['Address'].str.replace(' st ', ' street ')
df['Address'] = df['Address'].str.replace(' st$', ' street', regex=True)
df['Address'] = df['Address'].str.replace('st\.$', 'street', regex=True)
df['Address'] = df['Address'].str.replace(' st\. ', ' street ', regex=True)
df['Address'] = df['Address'].str.replace('st', 'st')
df['Address'] = df['Address'].str.replace('^w ', 'west ', regex=True)
df['Address'] = df['Address'].str.replace(' w\.', ' west', regex=True)
df['Address'] = df['Address'].str.replace(' w$', ' west', regex=True)
df['Address'] = df['Address'].str.replace('^e. ', 'east ', regex=True)
df['Address'] = df['Address'].str.replace('^e ', 'east ', regex=True)
df['Address'] = df['Address'].str.replace('blvd.', 'boulevard', regex=True)
df['Address'] = df['Address'].str.replace('blvd', 'boulevard', regex=True)
df['Address'] = df['Address'].str.replace(' rd$', ' road', regex=True)
df['Address'] = df['Address'].str.replace(' rd\.', ' road ', regex=True)
df['Address'] = df['Address'].str.replace(' rd ', ' road ', regex=True)
df['Address'] = df['Address'].str.replace('saint', 'st', regex=True)
df['Address'] = df['Address'].str.replace(' pl$', ' place', regex=True)
df['Address'] = df['Address'].str.replace('pkwy', 'parkway', regex=True)
df['Address'] = df['Address'].str.replace(' dr$', ' drive', regex=True)
df['Address'] = df['Address'].str.replace('^s ', 'south ', regex=True)
df['Address'] = df['Address'].str.replace(' cir$', ' circle', regex=True)
df['Address'] = df['Address'].str.replace(' ter$', ' terrace', regex=True)
df['Address'] = df['Address'].str.replace(' expy$', ' expressway', regex=True)
df['Address'] = df['Address'].str.replace(' hwy$', ' highway', regex=True)
df['Address'] = df['Address'].str.replace('fifth', '5', regex=True)
df['Address'] = df['Address'].str.replace('eight', '8', regex=True)
```

```python
# Hardcode address to match for some important locations
df.loc[7, 'Address'] = 'american museum of natural hist'
df.loc[1, 'Address'] = 'national 9/11 memorial'
df.loc[33, 'Address'] = 'city hall'
df.loc[34, 'Address'] = 'city hall park'
df.loc[73, 'Address'] = 'madison square garden'
df.loc[91, 'Address'] = 'yankee stadium'
df.loc[99, 'Address'] = 'poe park'
```

```python
colspecs = [(2, 34), (35, 36), (36, 37), (37, 42)]

# Read the file
sc = pd.read_fwf('../snd24Bcow.txt', colspecs=colspecs, header=None, names=["address_name", "f", "borough", "street_code"], dtype='string')
sc['address_name'] = sc['address_name'].str.lower()

sc.loc[sc.borough == "1", "borough"] = 'NY'
sc.loc[sc.borough == "2", "borough"] = 'BX'
sc.loc[sc.borough == "3", "borough"] = 'K'
sc.loc[sc.borough == "4", "borough"] = 'Q'
sc.loc[sc.borough == "5", "borough"] = 'R'
```

```python
sc.address_name
```

```python
# Create a mask where 'address_name' contains both 'american' and 'museum'
mask = (sc['address_name'].str.contains('american', case=False, na=False) &
    sc['address_name'].str.contains('museum', case=False, na=False))

# Apply the mask to 'sc'
result = sc[mask]

# Print the result
print(result.address_name)
```

```python
from fuzzywuzzy import fuzz

def fuzzy_match(row, threshold=97):  # increase the threshold for a near-perfect match
    return fuzz.token_sort_ratio(row['Address'], row['address_name']) > threshold

df['Address'] = df['Address'].str.replace('^\s*', '', regex=True)
df['Address'] = df['Address'].str.replace('^\d+', '', regex=True)
# Replace multiple whitespaces with a single whitespace in 'Address'
df['Address'] = df['Address'].str.replace('\s+', ' ', regex=True)

merged = sc.merge(df, left_on='borough', right_on='borough', suffixes=('', '_y'))


mask = (merged['address_name'].str.contains('american', case=False, na=False) &
    merged['address_name'].str.contains('museum', case=False, na=False) &
    merged['Address'].str.contains('museum', case=False, na=False))

print(merged[mask])

# Replace multiple whitespaces with a single whitespace in 'address_name'
merged['address_name'] = merged['address_name'].str.replace('\s+', ' ', regex=True)

# Fuzzy match entry addreses. Not needed anymore. I used it to see what are the difference in the street so I can
# add the rules above
mask = merged.apply(fuzzy_match, axis=1)
```
