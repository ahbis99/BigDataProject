Here we tried to rename the 2024 stret names entries in a consistent way so they are usable. Sadly this was not easy and we gave up on it 

```python
import pandas as pd

df = pd.read_csv('../big_data/2024.csv', dtype='string', nrows=1000000)
```

```python
colspecs = [(2, 34), (35, 36), (36, 37), (37, 42)]

# Read the file
sc = pd.read_fwf('../snd24Bcow.txt', colspecs=colspecs, header=None, names=["address_name", "f", "borough", "street_code"], dtype='string')
```

```python
df['Street Name'] = df['Street Name'].str.lower()
sc['address_name'] = sc['address_name'].str.lower()

# Remove extra whitespaces between words in 'Street Name' and 'address_name'
df['Street Name'] = df['Street Name'].str.replace('\s+', ' ', regex=True)
sc['address_name'] = sc['address_name'].str.replace('\s+', ' ', regex=True)

# Remove suffixes like 'th', 'st', 'nd', 'rd' from 'Street Name'
df['Street Name'] = df['Street Name'].str.replace('(\d+)(st|nd|rd|th)', r'\1', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' ave. ', ' avenue ')
df['Street Name'] = df['Street Name'].str.replace(' ave ', ' avenue ')
df['Street Name'] = df['Street Name'].str.replace(' ave$', ' avenue', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' st ', ' street ')
df['Street Name'] = df['Street Name'].str.replace(' st$', ' street', regex=True)
df['Street Name'] = df['Street Name'].str.replace('st.', 'st')
df['Street Name'] = df['Street Name'].str.replace('^w ', 'west ', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' w$', ' west', regex=True)
df['Street Name'] = df['Street Name'].str.replace('^e ', 'east ', regex=True)
df['Street Name'] = df['Street Name'].str.replace('blvd', 'boulevard', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' rd$', ' road', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' rd ', ' road ', regex=True)
df['Street Name'] = df['Street Name'].str.replace('saint', 'st', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' pl$', ' place', regex=True)
df['Street Name'] = df['Street Name'].str.replace('pkwy', 'parkway', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' dr$', ' drive', regex=True)
df['Street Name'] = df['Street Name'].str.replace('^s ', 'south ', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' cir$', ' circle', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' ter$', ' terrace', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' expy$', ' expressway', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' hwy$', ' highway', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' p ', ' place ', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' n$', ' north', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' pk$', ' park', regex=True)
df['Street Name'] = df['Street Name'].str.replace('^pk ', 'park ', regex=True)
df['Street Name'] = df['Street Name'].str.replace(' ft ', ' fort ', regex=True)
df['Street Name'] = df['Street Name'].str.replace('$ft ', 'fort ', regex=True)
df['Street Name'] = df['Street Name'].str.replace('$ins ', 'fort ', regex=True)
df['Street Name'] = df['Street Name'].str.replace('plygn$', 'playground', regex=True)
df['Street Name'] = df['Street Name'].str.replace('plygrn$', 'playground', regex=True)
```
